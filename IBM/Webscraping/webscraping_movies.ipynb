```python
import pandas as pd
import sqlite3
import requests
from bs4 import BeautifulSoup

# 엔티티 초기화
## 필요 url
url = 'https://web.archive.org/web/20230902185655/https://en.everybodywiki.com/100_Most_Highly-Ranked_Films'
## db이름
db_name = 'Movies.db'
## 레코드 저장할 테이블 이름
table_name = 'Top_50'
## 레코드 저장할 csv 이름
csv_path = 'c:/Users/user/Desktop/seongho/git/study/IBM/Webscraping/top_50_films.csv'
df = pd.DataFrame(columns=['Average Rank','Film','Year'])
## 상위 50개 결과만 가져오기 위한 초기화 루프 카운터
count = 0

# 웹 스크래핑을 위한 웹페이지 로딩
## 전체 웹 페이지 HTML 문서로 로드
html_page = requests.get(url).text
## HTML 형식으로 텍스트 파싱 후 정보 추출
data = BeautifulSoup(html_page, 'html.parser')

# 필요한 정보 스크래핑
## 모든 테이블 본문
tables = data.find_all('tbody')
## 첫번쨰 테이블의 모든 행
rows = tables[0].find_all('tr')

# 행을 반복하여 필요한 데이터 추출
## 변수 rows의 내용 반복
for row in rows:
    ## 루프 카운터 확안하여 50개 항목으로 제한
    if count<50:
        ## 행의 모든 td 데이터 객체 추출하여 col에 저장
        col = row.find_all('td')
        ## col의 길이가 0인지 확인(현재 행에 데이터가 없는 경우 찾기)
        if len(col)!=0:
            ## 출력 기록을 위해 이전 생성한 df의 열과 동일한 키를 가진 딕셔너리 생성
            data_dict = {"Average Rank": col[0].contents[0],
                         "Film": col[1].contents[0],
                         "Year": col[2].contents[0]}
            ## 딕셔너리를 df로 변환 후 기존 df와 연결
            df1 = pd.DataFrame(data_dict, index=[0])
            df = pd.concat([df,df1], ignore_index=True)
            ## 루프카운터 증가
            count+=1
    else:
        ## 카운터가 50에 도달하면 루프종료
        break

# df내용 출력
print(df)

# 데이터 저장
## CSV 파이로 저장
df.to_csv(csv_path)

# 데이터베이스에 저장
## 데이터 베이스 초기화
conn = sqlite3.connect(db_name)
## df를 테이블로 저장
df.to_sql(table_name, conn, if_exists='replace',index=False)
## 연결 닫기
conn.close()
```
